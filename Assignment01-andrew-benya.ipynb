```python
def convert_char_to_int(char):
    if (char.isalpha):
        return ord(char)
    else:
        return ' '
```

```python
def encrypt_with_ceaser(plaintext,key):
    # a=0, b=1, .... 
    
    # method will take number or character as key
    if(isinstance(key,str)):
        key = key.lower()
        key = ord(key) - ord('a')

    # operate on lowercase characters (arbitrary choice)
    plaintext = plaintext.lower()

    plaintext_ascii = []
    ascii_char_encrpyted_with_overflow = []    
    encrypted_ascii = []

    # populate plain text list with ascii equivlent
    for char in plaintext:
        plaintext_ascii.append(convert_char_to_int(char))

    # add the key value to all the ascii values of the plain text
    ascii_char_encrpyted_with_overflow = [char+key for char in plaintext_ascii]


    # if ascii values exceed the max value, reduce them by 26 (26 characters in working alphabet)
    encrypted_ascii = [x-26 if x>122 else x for x in ascii_char_encrpyted_with_overflow] 


    # convert all ascii integers to strings for output
    encrypted_ascii_strings = [str(int) for int in encrypted_ascii]


    # join together all strings into a single string (extra)
    string = "".join(encrypted_ascii_strings)


    # the big single string is convert into an integer (extra)
    final_int = int(string)


    return encrypted_ascii_strings

    



```

### Fence Rail Cipher

```python
plainText = 'Let us Meet tomorrow at Two'
```

```python
def fence_rail_encryption(plaintext,key):
    # key determines depth of array
    plaintext = plaintext.upper().replace(' ','')

    # fill matrix with placeholders
    working_matrix = [['0' for x in range(len(plaintext))] for y in range(key)]
    
    numberOfRails = key-1
    plaintext_list = [ x for x in plaintext]
    

    j=0
    j_increasing = True
    for i in range(len(plaintext_list)):

        working_matrix[j][i] = plaintext_list[i]
        
        if j==numberOfRails:            
            j-=1
            j_increasing=False
        elif j==0:
            j+=1
            j_increasing=True
        elif j_increasing==True:
            j+=1
        elif j_increasing==False:
            j-=1
        


     
    new_word_list = []
    for item in working_matrix:
        new_word = [x for x in item if x!='0']
        new_word_list.append(new_word)

    new_string=''
    for item in new_word_list:
        for x in item:
            new_string += x
        new_string += ' '


    return(new_string)


            

```

## 2.	Compare and contrast the two algorithms. 


The caesar cipher is logically simplier than the fence rail encryption algorithm, but because it is a simple, iterative algorithm it is susecptable to hueristic attacks like letter frequency analysis.  The number of possible keys is limited to the size of the alphabet, therefore brute force attacks make this encrpytion trivial, if the method of encryption is known.   The fence rail algorithm also has a limited number of keys, but that limit is dependent on the length of the plain text.  Fence rail also transposes its characters making hueristic analysis less useful, but not moot.  One could still analyze letter frequency successfully, but characters are moved out of position so the order of the characters changes.  For example, in fence rail encryption double letters will not be ciphered together in order.
