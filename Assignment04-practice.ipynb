```python
import pandas as pd
import numpy as np
```

```python
from sklearn.linear_model import *
from sklearn.tree import *
from sklearn.naive_bayes import *
from sklearn.neighbors import *
from sklearn.metrics import accuracy_score

from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plot_tree
```

```python
%matplotlib inline
```

```python
data = pd.read_csv("C:\\Users\\andre\\Documents\\BOWIE\\OneDrive - Bowie State\\645-OneDrive\\network-logs.csv")
```

```python
samples = data.iloc[:, [1,2]].values
targets = data['ANOMALY'].values
```

```python
print(samples)
print(targets)
```

```python
samples.shape
```

```python
targets.shape
```

```python
training_samples, testing_samples, training_targets, testing_targets = train_test_split(
    samples, targets, test_size=0.3,random_state=0)
```

```python
training_samples.shape
```

```python
testing_samples.shape
```

```python
knc = KNeighborsClassifier(n_neighbors=2)
knc.fit(training_samples, training_targets)
knc_prediction = knc.predict(testing_samples)
knc_accuracy = 100.0 * accuracy_score(testing_targets, knc_prediction)
print(f"K-Nearest Neighbors accuracy: {str(knc_accuracy)}")
```

```python
# decision tree
dtc = DecisionTreeClassifier(random_state=0)
dtc.fit(training_samples,training_targets)
dtc_prediction = dtc.predict(testing_samples)
dtc_accuracy = 100.0 * accuracy_score(testing_targets, dtc_prediction)
print(f"Decision Tree Accuracy: {str(dtc_accuracy)}")
```

```python
# Gaussian Naive Bayes
gnb = GaussianNB()
gnb.fit(training_samples,training_targets)
gnb_prediction = dtc.predict(testing_samples)
gnb_accuracy = 100.0 * accuracy_score(testing_targets, gnb_prediction)
print(f"Gauss Accuracy: {str(gnb_accuracy)}")
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
clf = MLPClassifier(solver='adam', alpha=1e-5,
    hidden_layer_sizes=(100,10,2), random_state=1, activation='relu')
```

```python
clf.fit(training_samples,training_targets)
clf_prediction = clf.predict(testing_samples)
clf_accuracy = 100.0 * accuracy_score(testing_targets, clf_prediction)
```

```python


print(f"K-Nearest Neighbors accuracy: {str(knc_accuracy)}")
print(f"Decision Tree Accuracy: {str(dtc_accuracy)}")
print(f"Gauss Accuracy: {str(gnb_accuracy)}")
print(f"CLF  Accuracy: {str(clf_accuracy)}")
```
