```python
key = b'mysecretpassword' # 16 byte password
cipher = AES.new(key, AES.MODE_CBC)

plaintext1 = b'plaintext goes here'

plaintext2 = b'plaintext goes hera'




print(plaintext1)
print(plaintext2)
```

```python


ciphertext1 = cipher.encrypt(pad(plaintext1,AES.block_size))

ciphertext2 = cipher.encrypt(pad(plaintext2,AES.block_size))
```

```python
print(f'First ciphertext is {ciphertext1}')
print(f'Second ciphertext is {ciphertext2}')
```

```python

iv = cipher.IV
#IV was auto generated
print(f'Initialization Vector is {iv}')
```

```python
cipher_to_decrypt = AES.new(key, AES.MODE_CBC, iv)
plaintext_decrpyt = unpad(cipher_to_decrypt.decrypt(ciphertext1), AES.block_size)
```

```python
plaintext_decrpyt.decode()
```

The AES cipher uses blocksizes of 128 bytes, so padding is required to perform the encrpytion.  
The cipher also relies on an initalization vector that serves as the first step in processing the input.  Here the Initialization vector is randomly generated.
If not for this initalization vector, if one plaintext block were to be identical to another their generated ciphertext block would be identical and would be that much easier to decrpyt by someone without key.
